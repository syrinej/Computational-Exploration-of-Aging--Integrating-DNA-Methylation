```python
import pandas as pd

# Load the CSV file containing GSM IDs and ages
sample_df = pd.read_csv('sample_df.csv')

# Display the first few rows to confirm it's loaded correctly
sample_df.head()
```

```python

# Keep only the first two columns from sample_df
sample_df = sample_df.iloc[:, :2]

# Display the resulting DataFrame
sample_df.head()
```

```python
# Get the total number of NaN values in the DataFrame
total_nans = df.isnull().sum().sum()
print(f"Total number of NaN values: {total_nans}")
```

```python
import pandas as pd

# Assuming your DataFrame is named sample_df
# Extract the number from the 'age' column using regex
sample_df['age'] = sample_df['age'].str.extract(r'(\d+)')

# Convert the 'age' column to integer type, if desired
sample_df['age'] = sample_df['age'].astype(int)

# Display the updated DataFrame
print(sample_df.head())
```

```python
sample_df.shape
```

```python
# Extract column names from df (assuming the first column is an ID or index column)
df_columns = df.columns[1:]  # Adjust if your DataFrame structure is different

# Extract sample IDs from sample_df
sample_ids = sample_df['sample']

# Find columns in df that are not in sample_df['sample']
missing_columns = [col for col in df_columns if col not in sample_ids.values]

# Display the missing columns
print("Columns in df that are not present in sample_df['sample']:")
print(missing_columns)
```

```python
# Ensure the first column is not included in the merge
df_no_id = df.set_index('ID_REF').T.reset_index().rename(columns={'index': 'sample'})

# Merge the dataframes on the sample IDs (GSM IDs)
merged_df = pd.merge(sample_df, df_no_id, on='sample')

# Display the first few rows of the merged dataframe
merged_df.head()
```

```python
merged_df.shape
```

```python
# Initialize an empty dictionary to store correlations
correlation_dict = {}

# Iterate over each CpG site (excluding the 'age' column)
for col in merged_df.columns:
    if col != 'age' and col != 'sample':  # Exclude 'age' and 'sample' columns from the correlation calculation
        correlation_dict[col] = merged_df[col].corr(merged_df['age'])

# Convert the dictionary to a Series for easy manipulation
correlations = pd.Series(correlation_dict)

# Display the top 10 CpG sites with the strongest positive and negative correlations with age
print("Top 10 CpG sites with strongest positive correlation with age:")
print(correlations.sort_values(ascending=False).head(10))

print("\nTop 10 CpG sites with strongest negative correlation with age:")
print(correlations.sort_values().head(10))

```

```python
import matplotlib.pyplot as plt

# Get the CpG site with the strongest positive correlation
strongest_positive_cpg = correlations.idxmax()

# Scatter plot for the strongest positively correlated CpG site
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['age'], merged_df[strongest_positive_cpg], alpha=0.5)
plt.title(f'Methylation Level of {strongest_positive_cpg} vs. Age')
plt.xlabel('Age')
plt.ylabel(f'{strongest_positive_cpg} Methylation Level')
plt.show()


```

```python
# Get the CpG site with the strongest negative correlation
strongest_negative_cpg = correlations.idxmin()

# Scatter plot for the strongest negatively correlated CpG site
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['age'], merged_df[strongest_negative_cpg], alpha=0.5)
plt.title(f'Methylation Level of {strongest_negative_cpg} vs. Age')
plt.xlabel('Age')
plt.ylabel(f'{strongest_negative_cpg} Methylation Level')
plt.show()


```

```python
import plotly.express as px
# Assuming 'correlations' and 'merged_df' are already defined from previous steps

# Select top 5 positively and negatively correlated CpG sites
top_positive_cpgs = correlations.nlargest(5).index
top_negative_cpgs = correlations.nsmallest(5).index

# Combine the top CpG sites
top_cpgs = top_positive_cpgs.union(top_negative_cpgs)

# Prepare data in long format for Plotly
top_cpgs_long = merged_df.melt(
    id_vars=['age'],
    value_vars=top_cpgs,
    var_name='CpG Site',
    value_name='Methylation Level'
)

# Create the interactive scatter plot
fig = px.scatter(
    top_cpgs_long,
    x='age',
    y='Methylation Level',
    color='CpG Site',
    title='Methylation Levels of Top 10 Correlated CpG Sites vs. Age',
    labels={
        'age': 'Age',
        'Methylation Level': 'Methylation Level',
        'CpG Site': 'CpG Site'
    },
    width=1000,   # Adjusts the width of the plot
    height=600    # Adjusts the height of the plot
)

# Update marker size and opacity
fig.update_traces(
    marker=dict(size=8, opacity=0.7),
    selector=dict(mode='markers')
)

# Update layout for better visualization
fig.update_layout(
    font=dict(size=14),
    legend_title_font=dict(size=16),
    title_font=dict(size=20),
    legend=dict(
        itemsizing='constant',
        title='CpG Sites',
        font=dict(size=12),
        orientation='h',
        x=0,
        y=1.1
    )
)

# Show the interactive plot
fig.show()

```

```python
import plotly.express as px

# Create age bins for better grouping in the violin plot
merged_df['age_group'] = pd.cut(
    merged_df['age'], 
    bins=[20, 40, 60, 80, 100], 
    labels=['20-40', '40-60', '60-80', '80-100']
)

# Drop rows where 'age_group' is NaN
merged_df = merged_df.dropna(subset=['age_group'])

# Select the top 5 CpG sites (either positive or negative correlations)
top_cpgs = correlations.nlargest(5).index

# Reshape the DataFrame for use with Plotly Express
top_cpgs_long = merged_df.melt(
    id_vars=['age_group'],
    value_vars=top_cpgs,
    var_name='CpG Site',
    value_name='Methylation Level'
)

# Customize the color palette for age groups
age_group_colors = ['#FF6347', '#32CD32', '#1E90FF', '#FFD700']

# Create the violin plot with facets for each CpG site
fig = px.violin(
    top_cpgs_long,
    x='age_group',
    y='Methylation Level',
    color='age_group',
    facet_col='CpG Site',  # Create a facet for each CpG site
    box=True,
    points='all',
    color_discrete_sequence=age_group_colors,
    title="Violin Plots of Methylation Levels for Top 5 CpG Sites Across Age Groups",
    width=1200,
    height=600
)

# Show the plot
fig.show()


```

```python
# Select the top CpG sites based on their correlation with age
top_cpg_sites = correlations.abs().sort_values(ascending=False).head(100).index  # Adjust the number based on your needs

# Prepare the data for modeling
X = merged_df[top_cpg_sites]  # Features: selected CpG sites
y = merged_df['age']          # Target: chronological age
```

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import cross_val_score

# Initialize the LASSO model with cross-validation
model = LassoCV(cv=5, random_state=42)

# Train the model
model.fit(X, y)

# Evaluate the model using cross-validation
scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error')
print(f'Mean Absolute Error: {-scores.mean()}')

# Display the selected features (CpG sites) used in the model
selected_features = X.columns[model.coef_ != 0]
print(f'Selected CpG Sites: {selected_features}')
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Predict the biological age on the training data
y_pred = model.predict(X)

# Calculate performance metrics
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')

# Scatter plot of predicted age vs. chronological age
plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred, alpha=0.5)
plt.xlabel('Chronological Age')
plt.ylabel('Predicted Biological Age')
plt.title('Predicted vs. Actual Age using Epigenetic Clock')
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')  # Diagonal line for reference
plt.show()

# Plot the coefficients (importance) of selected CpG sites
plt.figure(figsize=(12, 8))
plt.barh(selected_features, model.coef_[model.coef_ != 0])
plt.xlabel('Coefficient Value')
plt.ylabel('CpG Site')
plt.title('Importance of Selected CpG Sites in Epigenetic Clock')
plt.show()
```
